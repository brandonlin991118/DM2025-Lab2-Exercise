# **Competition Code**

## 1. Preprocessing Steps

````python
### Add the code related to the preprocessing steps in cells inside this section
# ============================================================

import pandas as pd
import numpy as np
import re
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# -------------------------------
# Load the dataset files
# -------------------------------
base_path = "/kaggle/input/dm-lab-2-private-competition"

posts = pd.read_json(os.path.join(base_path, "final_posts.json"), lines=False)
ids = pd.read_csv(os.path.join(base_path, "data_identification.csv"))
labels = pd.read_csv(os.path.join(base_path, "emotion.csv"))

print("‚úÖ Loaded all files successfully.")
print("posts:", posts.shape)
print("ids:", ids.shape)
print("labels:", labels.shape)

# -------------------------------
# Flatten JSON if needed
# -------------------------------
if "root" in posts.columns:
    print("üß© Flattening nested JSON...")
    posts = pd.json_normalize(posts["root"])

# Detect the post ID column automatically
possible_keys = [
    "id", "post_id", "tweet_id", "Id", "_source.post.post_id"
]
key_col = next((k for k in possible_keys if k in posts.columns), None)

if key_col is None:
    raise KeyError(f"‚ùå Could not find an ID column. Found: {posts.columns.tolist()}")

print(f"‚úÖ Detected ID column: '{key_col}'")

# Rename columns for consistency
posts.rename(columns={key_col: "post_id"}, inplace=True)
ids.rename(columns={"id": "post_id", "split": "data_type"}, inplace=True)
labels.rename(columns={"id": "post_id"}, inplace=True)

# -------------------------------
# Merge datasets
# -------------------------------
df = posts.merge(ids, on="post_id", how="left")
df = df.merge(labels, on="post_id", how="left")

print("‚úÖ Data merged successfully!")
print("Merged shape:", df.shape)

# -------------------------------
# Split into train/test
# -------------------------------
train_df = df[df["data_type"] == "train"].dropna(subset=["emotion"])
test_df = df[df["data_type"] == "test"]

print(f"üìä Train samples: {len(train_df)} | Test samples: {len(test_df)}")
````

## 2. Feature Engineering Steps

````python
### Add the code related to the feature engineering steps in cells inside this section
# ============================================================

from sklearn.feature_extraction.text import TfidfVectorizer

# -------------------------------
# Text cleaning
# -------------------------------
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Detect the correct text column
text_col = next((c for c in ["text", "_source.post.text"] if c in df.columns), None)
if text_col is None:
    raise KeyError(f"‚ùå No text column found! Found: {df.columns.tolist()}")

print(f"‚úÖ Using text column: '{text_col}'")

# Apply text cleaning
train_df.loc[:, "clean_text"] = train_df[text_col].apply(clean_text)
test_df.loc[:, "clean_text"] = test_df[text_col].apply(clean_text)

# -------------------------------
# Encode labels
# -------------------------------
le = LabelEncoder()
train_df.loc[:, "label"] = le.fit_transform(train_df["emotion"])
print("üé≠ Emotion Classes:", list(le.classes_))

# -------------------------------
# Split validation
# -------------------------------
X_train, X_val, y_train, y_val = train_test_split(
    train_df["clean_text"],
    train_df["label"],
    test_size=0.2,
    random_state=42,
    stratify=train_df["label"]
)
````

## 3. Model Implementation Steps

````python
### Add the code related to the model implementation steps in cells inside this section

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, f1_score

# -------------------------------
# Model definition
# -------------------------------
model = Pipeline([
    ("tfidf", TfidfVectorizer(
        max_features=25000,
        ngram_range=(1, 2),
        sublinear_tf=True
    )),
    ("clf", LogisticRegression(max_iter=300, n_jobs=-1))
])

# -------------------------------
# Training and validation
# -------------------------------
print("üöÄ Training model...")
model.fit(X_train, y_train)

y_pred = model.predict(X_val)
val_f1 = f1_score(y_val, y_pred, average="macro")

print("\nüìä Validation Results:")
print(f"Macro F1 Score: {val_f1:.4f}")
print(classification_report(y_val, y_pred, target_names=le.classes_))

# -------------------------------
# Retrain on full dataset
# -------------------------------
print("\nüîÅ Retraining on full data...")
model.fit(train_df["clean_text"], train_df["label"])

# -------------------------------
# Predict test set and save submission
# -------------------------------
test_preds = model.predict(test_df["clean_text"])
test_df.loc[:, "emotion"] = le.inverse_transform(test_preds)

submission = pd.DataFrame({
    "id": test_df["post_id"],
    "emotion": test_df["emotion"]
})
submission.to_csv("/kaggle/working/submission.csv", index=False)

print("\n‚úÖ submission.csv successfully created!")
display(submission.head())

print("\nüéØ Upload '/kaggle/working/submission.csv' to the Kaggle competition page.")
````
